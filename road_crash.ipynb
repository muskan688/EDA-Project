# Exploratory Data Analysis of Nigerian Road Traffic Crashes (2020-2024) .

- EDA helps to understand the data pattern, tell us which all informations are important and which all attributes are useless and which are useful.

Attributes
- Quarter - The quarter of the year when crashes were reported.
- State - The state in Nigeria where crashes occurred.
- Total_Crashes - Total number of crashes reported.
- Num_Injured - Number of individuals injured
- Num_Killed - Number of fatalities
- Total_Vehicles_Involved - Total vehicles involved in the cras
- SPV, DAD, PWR, FTQ - Different contributing factors or vehicle types (meanings need futher 
clarifcation
-  Other_Factors - Additional unspecified contributingfactorss

Data Source- https://www.kaggle.com/datasets/akinniyiakinwande/nigerian-traffic-crashes-2020-2024.

 # Univariate Analysis-- single column, categorical, numerical

```python
#Distribution of total crashes

plt.figure(figsize=(6,3)) #using histogram because its a numerical data
sns.histplot(data['Total_Crashes'], kde=True, color='green') #kde is line above the graph
plt.title('Distribution of Total Crashes')
plt.show()
```

```python
#count plot

plt.figure(figsize=(20,3)) 
sns.countplot(x='State', data=data, color='green') 
plt.title('Distribution of States')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(16,5))
orders= data['State'].value_counts().index
sns.countplot(x='State', data=data,order=orders, palette='viridis')
for p in plt.gca().patches:  #
    plt.gca().annotate(f'{p.get_height()}', (p.get_width()/2., p.get_height()),ha='center',va='baseline',fontsize=8,color='black', xytext=(0,5), textcoords='offset points')
    
plt.xticks(rotation=45, ha='right' )
plt.title('Distribution of States')
plt.show()
```

# Bivariate Analysis

```python
#state vs total crashes

plt.figure(figsize=(16,5))
sorted=data.sort_values(by='Total_Crashes', ascending=False).head(100)
sns.barplot(x='State',y='Total_Crashes',data=sorted,color='green')
plt.title('State vs Total Crashes')
plt.show()
```

```python
plt.figure(figsize=(16,5))
sorted=data.sort_values(by='Total_Crashes', ascending=True).head(100)
sns.barplot(x='State',y='Total_Crashes',data=sorted,color='green')
plt.title('State vs Total Crashes')
plt.show()
```

```python
#split the data quarter
data['Year']=data['Quarter'].apply(lambda x: x.split()[1])
data['Quarter']=data['Quarter'].apply(lambda x: x.split()[0])
data.head()
```

```python
top_states=data.groupby('State')['Total_Crashes'].sum().nlargest(10).index
top_10 =data[data['State'].isin(top_states)]

plt.figure(figsize=(16,5))
sns.lineplot(x='Year',y='Total_Crashes',data=top_10,hue='State',marker='o')
plt.title('State vs Total Crashes')
plt.show()
```

```python
top_states=data.groupby('State')['Total_Crashes'].sum().nlargest(10).index
top_10 =data[data['State'].isin(top_states)]

plt.figure(figsize=(16,5))
sns.lineplot(x='Quarter',y='Total_Crashes',data=top_10,hue='State',marker='o')
plt.title('State vs Total Crashes')
plt.show()
```

```python
#quarter vs total crashes
plt.figure(figsize=(16,5))
sns.boxplot(x='Quarter',y='Total_Crashes',data=data,hue='Quarter')
plt.title('Top 10 State vs Total Crashes')
plt.show()
```

```python
#quarter vs total crashes
plt.figure(figsize=(16,5))
sns.boxplot(x='Year',y='Total_Vehicles_Involved',data=data,hue='Year')
plt.title('Year vs Total Vehicles Involved')
plt.show()
```

```python
#quarter vs total crashes
plt.figure(figsize=(16,5))
sns.scatterplot(x='Total_Crashes',y='Num_Killed',data=data,hue='Year')
plt.title('Year vs Total crashes')
plt.show()
```

```python
#heat map
corr_rel= data[['Total_Crashes','SPV','DAD','FTQ','PWR','Other_Factors']].corr()
corr_data= corr_rel[['Total_Crashes']].loc[['SPV','DAD','FTQ','PWR','Other_Factors']]

plt.figure(figsize=(16,5))
sns.heatmap(corr_data, annot=True, cmap='coolwarm')
plt.title('Correlation')
plt.show()
```
